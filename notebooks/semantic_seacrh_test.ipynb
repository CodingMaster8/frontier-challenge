### Tool testing: Semantic Search Tool

```python
from frontier_challenge.tools import SemanticSearchTool
```

```python
queries = [
    "sustainable technology investing",
    "Bradesco gold fund",
    "emerging markets bonds",
    "Funds that invest in renewable energy",
    "high yield corporate debt",
]
```

```python
tool = SemanticSearchTool()
```

```python
print("\nBuilding vector index...")
count = tool.build_index()
print(f"Indexed {count:,} funds")
```

```python
for query in queries:
    print("\n" + "=" * 80)
    results, explanation = tool.search_with_explanation(query, top_k=5)
    print(explanation)
```
