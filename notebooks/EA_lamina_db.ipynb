### Exploratory Analysis of Lamina tables in the database

```python
import duckdb
import pandas as pd
```

```python
conn = duckdb.connect('../data/br_funds.db')

tables = conn.execute("SHOW TABLES").fetchall()
print("Available tables:", tables)

conn.close()
```

### lamina_carteira table
- 47,879 rows and 7 columns
- TP_ATIVO : Type of asset held by the fund
- PR_PL_ATIVO : Percentage of the fund's net worth invested in each asset type
- DT_COMPTC	: Date of the report

```python
conn = duckdb.connect('../data/br_funds.db')

df = conn.execute("SELECT * FROM lamina_carteira LIMIT 20").fetchdf()

conn.close()

df.head()
```
